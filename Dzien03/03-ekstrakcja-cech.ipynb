```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
import seaborn as sns
```

```python
X , y =load_breast_cancer(return_X_y=True, as_frame=True)
```

```python
df = pd.concat([X,y], axis=1)
df
```

```python
plt.figure(figsize=(18,18))
sns.heatmap( df.corr(), annot=True, cmap='jet' )
```

```python
N_FEAT = 5
best_features = {}
```

```python
best_columns = df.corr().iloc[-1].abs().sort_values(ascending=False).iloc[1:N_FEAT+1].index
best_features["corr"] = best_columns
best_columns
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
# estymator na 5 kolumnach wyciagnietych z wykorzystaniem korelacji
X_train, X_test, y_train, y_test = train_test_split( df[best_columns], df.target, test_size=.2, random_state=0 )
model = KNeighborsClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
pd.DataFrame( confusion_matrix(y_test, model.predict(X_test)) )
```

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
# estymator na podstawie cech wybranych klasÄ… SelectKBest
fe = SelectKBest(chi2, k=N_FEAT)
fe.fit(X, y)
best_columns = X.iloc[ : , fe.get_support()].columns.tolist()
print(best_columns)
best_features["select_k_best"] = best_columns

X_train, X_test, y_train, y_test = train_test_split( df[best_columns], df.target, test_size=.2, random_state=0 )
model = KNeighborsClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
pd.DataFrame( confusion_matrix(y_test, model.predict(X_test)) )
```

```python
# Technika RFE
from sklearn.feature_selection import RFE
```

```python
fe = RFE( DecisionTreeClassifier(random_state=0) , n_features_to_select=N_FEAT )
fe.fit(X, y)
best_columns = X.iloc[ : , fe.get_support()].columns.tolist()
print(best_columns)
best_features["rfe"] = best_columns

X_train, X_test, y_train, y_test = train_test_split( df[best_columns], df.target, test_size=.2, random_state=0 )
model = KNeighborsClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
pd.DataFrame( confusion_matrix(y_test, model.predict(X_test)) )
```

```python
# Technika opakowywania
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
```

```python
#fe = RFE( DecisionTreeClassifier(random_state=0) , n_features_to_select=N_FEAT )
fe = SelectFromModel( RandomForestClassifier(random_state=0) , max_features=N_FEAT)
fe.fit(X, y)
best_columns = X.iloc[ : , fe.get_support()].columns.tolist()
print(best_columns)
best_features["select_from_model"] = best_columns

X_train, X_test, y_train, y_test = train_test_split( df[best_columns], df.target, test_size=.2, random_state=0 )
model = KNeighborsClassifier()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
pd.DataFrame( confusion_matrix(y_test, model.predict(X_test)) )
```
