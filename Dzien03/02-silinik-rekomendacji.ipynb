```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sentences = ["Julia kocha Marka", "Marek kocha Annę", "Anna i Julia kochają Marka", "Dzisiaj świeci słońce"]
```

```python
cv = CountVectorizer()
wm = cv.fit_transform(sentences)
```

```python
pd.DataFrame( wm.toarray(), columns=cv.get_feature_names() )
```

```python
pd.DataFrame( cosine_similarity(wm) )
```

```python
df = pd.read_csv("http://51.91.120.89/extras/movie_dataset.csv")
df.head(3)
```

```python
df.columns
```

```python
# wybór cech na potrzeby rekomendacji
features = ['genres', 'keywords',  'title', 'cast',  'director']
```

```python
df.isna().sum()
```

```python
for f in features:
    df[f].fillna("", inplace=True)
```

```python
df["recommend"]=df.genres+" "+df.keywords+" "+df.title+" "+df.cast+" "+df.director
```

```python
df.recommend.iloc[123]
```

```python
cv = CountVectorizer()
wm = cv.fit_transform(df.recommend)
wm.shape
```

```python
similarity = cosine_similarity(wm)
similarity.shape
```

```python
sorted( list( similarity[123] ) , reverse=True)
```

```python
def get_title_by_index(ix):
    return df.iloc[ix].title
```

```python
get_title_by_index(1458)
```

```python
for mi, sim in sorted( list( enumerate(similarity[1458]) ) , reverse=True, key=lambda x:x[1])[1:11]:
    print(get_title_by_index(mi), round(sim*100,1) )
```
