```python
import pandas as pd
```

```python
df = pd.read_csv("http://51.91.120.89/extras/advertising.csv", usecols=["TV","sales"])
df
```

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
X = df.TV.values.reshape(-1,1)
y = df.sales.values
```

```python
model = DecisionTreeRegressor(max_depth=7, random_state=0)
model.fit(X,y)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X, y)

x = np.unique(X).reshape(-1,1)
y_pred = model.predict(x)
plt.plot(x, y_pred, c="r")
print( r2_score(y, model.predict(X) ))
```

```python
df = pd.read_csv("http://51.91.120.89/extras/otomoto-dt.csv")
df
```

```python
df_B8.drop(columns=['Unnamed: 0', 'ID','Marka pojazdu', 'Model pojazdu','Wersja', 'WyposaÅ¼enie','Kategoria', ] , inplace=True)
```

```python
df_B8 = pd.get_dummies(df_B8)
```

```python
df_B8.columns
```

```python
plt.hist( df_B8.Cena[df_B8.Cena<80_000], bins=20 );
```

```python
X = df_B8.iloc[: , 1:]
y = df_B8.Cena
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
for i in range(2,15):
    model = DecisionTreeRegressor(max_depth=i, random_state=0)
    model.fit(X_train, y_train)
    print(f"DT={i}, score={model.score(X_test, y_test)}")
```

```python
i = 5
model = DecisionTreeRegressor(max_depth=i, random_state=0)
model.fit(X_train, y_train)
print(f"DT={i}, score={model.score(X_test, y_test)}")
```

```python
pd.DataFrame( model.feature_importances_, X.columns).sort_values(0, ascending=False)
```
