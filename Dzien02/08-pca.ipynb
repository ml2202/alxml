```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
import seaborn as sns
```

```python
ds = load_breast_cancer()
```

```python
print(ds.DESCR)
```

```python
df = pd.DataFrame(ds.data, columns=ds.feature_names)
df
```

```python
# import klasy odpowiedzialnej za PCA
from sklearn.decomposition import PCA
```

```python
# redukcja wymiarowosci
N_DIM = 5
pca = PCA(n_components=N_DIM, random_state=0)
X_pca = pca.fit_transform(df)
```

```python
X_pca.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
%%timeit -r 10 -n 100

# estymator na danych oryginalnych

X_train, X_test, y_train, y_test = train_test_split(df, ds.target, test_size=0.2, random_state=0)

model = DecisionTreeClassifier(random_state=0, max_depth=7)
model.fit(X_train, y_train)
#print(model.score(X_test, y_test))
#pd.DataFrame( confusion_matrix(y_test, model.predict(X_test) ) )
```

```python
%%timeit -r 10 -n 100

# estymator na PCA

X_train, X_test, y_train, y_test = train_test_split(X_pca, ds.target, test_size=0.2, random_state=0)

model = DecisionTreeClassifier(random_state=0, max_depth=7)
model.fit(X_train, y_train)
#print(model.score(X_test, y_test))
#pd.DataFrame( confusion_matrix(y_test, model.predict(X_test) ) )
```
